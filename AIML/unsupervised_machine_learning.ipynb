```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('clustering_data_1.csv.xls')
```

```python
df.head()
```

```python
plt.scatter(x='X',y='Y',data=df,marker='o')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
kmeans=KMeans(n_clusters=2)
kmeans.fit(df)
```

```python
cluster_assingment=kmeans.predict(df)
```

```python
cluster_assingment
```

```python
sns.scatterplot(x='X',y='Y',data=df,hue=cluster_assingment)
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
cluster_assignment=kmeans.predict(df)
```

```python
sns.scatterplot(x='X',y='Y',data=df,hue=cluster_assignment)
```

```python
y_predict=kmeans.fit_predict(df[['X','Y']])
y_predict
```

```python
df['cluster_prediction']=y_predict
df.head()
```

```python
df1=df[df['cluster_prediction']==0]
df2=df[df['cluster_prediction']==1]
df3=df[df['cluster_prediction']==2]
```

```python
plt.scatter(df1.X,df1['Y'],color='green',label='cluster1')
plt.scatter(df2.X,df2['Y'],color='red',label='cluster2')
plt.scatter(df3.X,df3['Y'],color='black',label='cluster3')
plt.legend()
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(df1.X,df1['Y'],color='grey',label='cluster1')
plt.scatter(df2.X,df2['Y'],color='lightblue',label='cluster2')
plt.scatter(df3.X,df3['Y'],color='yellow',label='cluster3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black',marker='*',label='centroid')
plt.legend()
plt.show()
```
