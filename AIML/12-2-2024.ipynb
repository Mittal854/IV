## <font color="red">Visulation of data in Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tp=sns.load_dataset('tips')
```

```python
tp.head()
```

```python
tp.info()
```

```python
tp.columns
```

```python
# import warnings
# warnings.filterwarnings('ignore')
sns.relplot(x='total_bill',y='tip',data=tp)
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex', data=tp)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker', data=tp)
```

```python
color=['red','black']
sns.relplot(x='total_bill',y='tip',hue='smoker', data=tp,palette=color)
```

```python
sns.set(style='darkgrid')
color=['red','black']
sns.relplot(x='total_bill',y='tip',hue='smoker', data=tp,palette=color)
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex', data=tp,style='time')
```

```python
tp['time'].value_counts()
```

```python
sns.countplot(x='time' ,data=tp)
```
