```python
def even_odd(x):
    if(x % 2 == 0):
        return "even"
    else:
        return "odd"
```

```python
even_odd(2)
```

```python
print(list(map(even_odd, [1, 2, 3, 4, 5])))
```

```python
list(map(even_odd, [1, 2, 3, 4, 5]))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x=np.arange(0,np.pi*4,0.1)
x
```

```python
y1=np.sin(x)
y1
```

```python
y2=np.cos(x)
y2
```

```python
y3=np.tan(x)
y3
```

```python
fig=plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.title('sin(x)')
plt.plot(x,y1)
plt.subplot(1,3,2)
plt.title('cos(x)')
plt.plot(x,y2)
plt.subplot(1,3,3)
plt.title('tan(x)')
plt.plot(x,y3)
```
