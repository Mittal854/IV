# Sigmod Function:-
## The Logistic function in the supervised machine learning is a type of a sigmod, a class of function with same specific properties. Sigmod is a mathematical function that takes any real number and maps it to a probability between 0 and 1

## y=1/(1+e(-x))

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmod(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-5,5,1000)
y=sigmod(x)
plt.plot(x,y,label='sigmod')
plt.legend()
plt.grid()
plt.show()
```
